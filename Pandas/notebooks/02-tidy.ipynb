```python
billboard = pd.read_csv('../data/billboard.csv')
```

```python
billboard.head()
```

```python
billboard.melt(id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
               value_name='rank',
               var_name='week');
```

```python
# Wrap it if you are writing a big expression im multiple lines or use "\".
(
billboard
  .melt(id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
        value_name='rank',
        var_name='week')
  .groupby('artist')['rank']
  .mean().sort_values()
);
```

```python
billboard.melt\
(id_vars=['year', 'artist', 'track', 'time', 'date.entered'],
        value_name='rank',
        var_name='week')\
  .groupby('artist')['rank']\
  .mean();
```

# Tranforming a weather Date in rows and coloumns

```python
weather = pd.read_csv('../data/weather.csv')
```

```python
weather.head()
```

```python
weather_long = weather.melt(id_vars=['id', 'year', 'month', 'element'],
                            var_name='day',
                            value_name='temp')
```

```python
weather_long.head()
```

### Back to the previous state

```python
(weather_long.pivot_table(index=['id', 'year', 'month', 'day'],
                         columns='element',
                         values='temp')
.reset_index()
)
```
